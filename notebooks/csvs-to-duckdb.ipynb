```python
# Create a persistent DuckDB database
con = duckdb.connect(database='week1.duckdb', read_only=False)

# Load us-states-gdp-by-year.csv, note manual trimming to skip the first 3 rows and dropping the last 5 rows
con.execute('''
CREATE OR REPLACE TABLE us_states_gdp_by_year AS
SELECT *
FROM read_csv('../edx-power-bi-exercise-files/Week 1/us-states-gdp-by-year_trimed.csv', 
    header=true, 
    auto_detect=true)
''')

# Load us-states.csv
con.execute('''
    CREATE OR REPLACE TABLE us_states AS
    SELECT * FROM read_csv('../edx-power-bi-exercise-files/Week 1/us-states.csv')
''')

# Load us-states-population-by-year.csv
con.execute('''
    CREATE OR REPLACE TABLE us_states_population_by_year AS
    SELECT * FROM read_csv('../edx-power-bi-exercise-files/Week 1/us-states-population-by-year.csv')
''')

for column_name in ['AL','AK','AZ','AR','CA','CO','CT','DE','FL','GA','HI','ID','IL','IN','IA','KS','KY','LA','ME','MD','MA','MI','MN','MS','MO','MT','NE','NV','NH','NJ','NM','NY','NC','ND','OH','OK','OR','PA','RI','SC','SD','TN','TX','UT','VT','VA','WA','WV','WI','WY','DC']:
    con.execute(f"""
        UPDATE us_states_population_by_year
        SET {column_name} = CAST(REPLACE({column_name}, ',', '') AS INTEGER)
    """)

for column_name in ['AL','AK',...,'DC']:
    con.execute(f"""
        ALTER TABLE us_states_population_by_year
        ALTER {column_name} TYPE INTEGER
    """)

print('Tables loaded successfully into the persistent DuckDB database.')
```
